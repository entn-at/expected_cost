```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of posterior probabilities for class 1
posterior_probs = np.linspace(0.001, 0.999, 100)

fig, axs = plt.subplots(1,3, figsize=(8, 2.5))

# Cost
c12 = 1
c21 = 2
thr = c12 / (c12 + c21)
loss_class_2 = c21 * (posterior_probs<thr)
loss_class_1 = c12 * (posterior_probs>=thr)
ax = axs[0]
ax.plot(posterior_probs, loss_class_1, label='$h = H_1$')
ax.plot(posterior_probs, loss_class_2, label='$h = H_2$')
ax.set_xlabel('$q_2$')
ax.set_ylabel('Cost for $c_{12}=1$, $c_{21}=2$')

# Log loss
loss_class_2 = -1 * np.log(posterior_probs)
loss_class_1 = -1 * np.log(1-posterior_probs)
ax = axs[1]
ax.plot(posterior_probs, loss_class_1, label='$h = H_1$')
ax.plot(posterior_probs, loss_class_2, label='$h = H_2$')
ax.set_xlabel('$q_2$')
ax.set_ylabel('Log Loss')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.1f}'))

# Brier loss 
loss_class_2 = 0.5 * (1 - posterior_probs) ** 2
loss_class_1 = 0.5 * posterior_probs ** 2
ax = axs[2]
ax.plot(posterior_probs, loss_class_1, label='$h = H_1$')
ax.plot(posterior_probs, loss_class_2, label='$h = H_2$')
ax.set_xlabel('$q_2$')
ax.set_ylabel('Brier Loss')

# Add legend
ax.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))

plt.tight_layout()

plt.savefig('outputs/losses.pdf', bbox_inches='tight')
```
